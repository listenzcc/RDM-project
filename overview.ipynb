```python
# Build data frame
df = resource_table_input_data.query('extension == "jpg"').copy()
for type in df['module'].unique():
    for image in df['set'].unique():
        print(type, image)

# Draw the example image
index = np.array(df.index)
np.random.shuffle(index)
df = df.loc[index]
images = df.groupby(by='set').first()
n = len(images)
fig, axes = plt.subplots(1, n, figsize=(8, 8 / n))

for j, s in enumerate(images.index):
    img = Image.open(images.loc[s, 'fullPath'])
    name = images.loc[s, 'name']
    ax = axes[j]
    ax.imshow(img)
    ax.set_title('{} - {}'.format(s, name))
    # ax.axis('off')

fig.suptitle('Example image')
plt.tight_layout()
plt.show()

jpg = df.loc[sorted(index)]
jpg
```
